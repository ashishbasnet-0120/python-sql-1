```python

#   another example

class Hello:
    def __init__(self):
        self.pin=''
        self.balance=0
        self.menu()




    def menu(self):
        user_desire=int(input("""
        Hello you want these services .choose wisely according to yourrequirement:
        1.press 1 to create pin
        2.press 2 to change pin
        3.press 3 for check balance
        4.press 4 for withdrawl money
        5.press 0  for exit 
        
        """))



        
        if user_desire==1:
            # create()
            self.create_pin()

        elif user_desire==2:
            # change pin()
            self.change_pin()

        elif user_desire==3:
            # check_balance()
            self.check_balance()

        elif user_desire==4:
            # withdrawmoney()
            self.withdraw()

        elif user_desire==0:
            # exit()
            self.exit()
            



    def create_pin(self):
        user_pin=input('create your pin:')
        self.pin=user_pin


        balance_amount=int(input("enter the amount of balance you want to deposit:"))
        self.balance=balance_amount
        print('your pin has been successfully created>')
        self.menu() # for continuity of program

    def change_pin(self):
        old_pin=input('enter your existing or old pin:')
        if old_pin==self.pin:
            new_pin=input("change your pin:")
            self.pin=new_pin

        else:
            print("sorry, your old pin doesn't match")
        self.menu()


    def check_balance(self):
        print("your balance is:",self.balance)
        self.menu()


    def withdraw(self):
        cash=int(input("enter the amount of money you want to withdraw:"))
        if cash>self.balance:
            print("your current balance is less than you proceeed")
        elif cash==self.balance:
            print("""
            you can't withdraw all the money.
            please withdraw less than your actual money
            """)
        else:
            new_balance=self.balance-cash
            self.balance=new_balance
        self.menu()

    def exit(self):
        print("thanks for visiting us ")
        

    





            
            
            
            
obj4=Hello()

# in above code using calling the function in constructor so it automatically executed 



```
