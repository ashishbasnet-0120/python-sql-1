```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python



iris=pd.read_csv(r"C:\Users\acer\Python-Sql-1\numpy_pandas_matplotlib_seaborn\Matplotlib and Seaborn\dataset_matplotlib\iris.csv")
iris.head()
```

```python



# colored scattered plots

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
plt.xlabel('sepal length ')
plt.ylabel('petal length')
plt.title('sepal vs petal')
plt.show()
```

```python


# colring each points on the basis of species 

iris.sample(5)
```

```python


iris['Species']=iris['Species'].replace({'Iris-setosa':0,'Iris-virginica':1,'Iris-versicolor':3})
iris.sample()
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter')
plt.xlabel('sepal length ')
plt.ylabel('petal length')
plt.title('sepal vs petal')
plt.show()
```

```python
# colormap function


plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter')
plt.xlabel('sepal length ')
plt.ylabel('petal length')
plt.title('sepal vs petal')
plt.colorbar()
plt.show()
```

```python
# alpha 


plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.5)
plt.xlabel('sepal length ')
plt.ylabel('petal length')
plt.title('sepal vs petal')
plt.colorbar()
plt.show()
```

```python




# plotsize

plt.figure(figsize=(15,7))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter')
plt.xlabel('sepal length ')
plt.ylabel('petal length')
plt.title('sepal vs petal')
plt.colorbar()

plt.show()
```

```python



# 3d surface plot  

# most imp in ml



x=np.linspace(-10,10,100)  # from -10 to 10 with 100  equidistance 
y=np.linspace(-10,10,100)


x.shape
```

```python
xx,yy=np.meshgrid(x,y)
xx.shape
```

```python
z=xx**2+yy**2
z.shape
```

```python


fig=plt.figure(figsize=(16,9))
ax=plt.subplot(projection="3d")
ax.plot_surface(xx,yy,z,cmap="viridis")
fig.colorbar()
```

```python



# need to remember 

fig=plt.figure(figsize=(16,9))
ax=plt.subplot(projection="3d")
p=ax.plot_surface(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```

```python

# contour plot


fig=plt.figure(figsize=(16,9))
ax=plt.subplot()
p=ax.contour(xx,yy,z,cmap="viridis")
fig.colorbar(p)

```

```python



# countour f
fig=plt.figure(figsize=(16,9))
ax=plt.subplot()
p=ax.contourf(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```

```python



# heatmap

ipl=pd.read_csv(r"C:\Users\acer\Python-Sql-1\numpy_pandas_matplotlib_seaborn\Matplotlib and Seaborn\dataset_matplotlib\IPL_Ball_by_Ball_2008_2022.csv")
ipl.sample(20)
```

```python


temp=ipl[(ipl['ballnumber'].isin([1,2,3,4,5,6])) & (ipl['batsman_run']==6)]

temp.head()
```

```python



df=temp.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')

plt.figure(figsize=(15,7))
plt.imshow(df)
plt.colorbar()
```
