```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
iris=sns.load_dataset('iris')
```

### categorical scatter plot

```python
# stripplot-->  axes level function

# catplot() --->   figure level function 


sns.stripplot(data=tips,x='day', y='tip')
```

```python

sns.catplot(data=tips,x='day', y='tip',kind='strip',jitter=False)
```

```python
sns.catplot(data=tips,x='day', y='tip',kind='strip',jitter=0.2)
```

```python
sns.catplot(data=tips,x='day', y='tip',kind='strip', hue='sex')
```

```python


# swarmplot
sns.catplot(data=tips,x='day', y='tip',kind='swarm', hue='sex')
```




# boxplot

```python
  # for single column

sns.boxplot(data=tips,y='total_bill')
```

```python
tips
```

```python
# for comparing boxplots or for categorical column as well


sns.catplot(data=tips,x='day',y='total_bill',kind='box')
```

```python
# hue parameter

sns.catplot(data=tips,x='day',y='total_bill',kind='box',hue='sex')
```




# violinplot

```python
sns.violinplot(data=tips, x='day',y='total_bill')
```

```python


sns.catplot(data=tips, x='day',y='total_bill',kind='violin',hue='sex')
```

```python

# split

sns.catplot(data=tips, x='day',y='total_bill',kind='violin',hue='sex',split=True)
```



## barplot

#### by default estimator is mean we can change it 

```python
sns.barplot(data=tips,x='sex',y='total_bill')
```

```python

# error bar is created when the x axis values gets repeated

# removing errorbar

sns.barplot(data=tips,x='sex',y='total_bill',estimator='median',errorbar=None)
```

```python



# hue
sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker')
```

```python



sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker',estimator="var")
```






### point plot 

##### shows the average (or other estimate) of a numeric variable for different categories using points instead of bars.

```python


# using figure level function

sns.catplot(data=tips,x='sex',y='total_bill',estimator='median',kind='point')
```

```python


# using hue
sns.catplot(data=tips,x='sex',y='total_bill',estimator='median',kind='point',hue='smoker')
```

```python
# removing errorbars
sns.catplot(data=tips,x='sex',y='total_bill',estimator='median',kind='point',errorbar=None)
```

```python
sns.catplot(data=tips,x='sex',y='total_bill',estimator='median',kind='point',errorbar=None,hue='smoker')
```
