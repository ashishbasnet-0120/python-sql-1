### relational plot

```python
## scatter plot
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python


tips=sns.load_dataset('tips')
tips
```

```python
# scatter plot ---> axes level function


sns.scatterplot(data=tips,x='tip',y='total_bill')


```

```python


# relpot---> figure level function  ---> square shape

sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter')
```

```python


# hue parameter
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex')
```

```python


sns.scatterplot(data=tips,x='tip',y='total_bill',hue='sex')
```

```python


# style paarameter
sns.scatterplot(data=tips,x='tip',y='total_bill',hue='sex',style='time')
```

```python
# size
sns.scatterplot(data=tips,x='tip',y='total_bill',hue='sex',style='time',size='size')
```



### matrix plot

```python
##  heatmap is a axes level function there is  no figure level function 







a
```

```python
df=a.pivot(index='country',columns='year',values='lifeExp')
df
```

```python
sns.heatmap(df)
```

```python


# since heatmap is a axes level function we can resize same what we did in matplotlib


plt.figure(figsize=(15,15))
sns.heatmap(df)
```

```python



asia=a[a['continent']=="Asia"]
asia
```

```python

new=asia.pivot(index='country',columns='year',values='lifeExp')
new

```

```python

plt.figure(figsize=(20,20))
sns.heatmap(new)
```

```python

# annot 
plt.figure(figsize=(15,15))
sns.heatmap(new,annot=True)
```

```python

# linewidth
plt.figure(figsize=(15,15))
sns.heatmap(new,annot=True,linewidth=1)
```

```python
#cmap
plt.figure(figsize=(15,15))
sns.heatmap(new,annot=True,linewidth=1,cmap='viridis')
```
