```python




# boolean indexing


a=np.random.randint(1,200,50).reshape(25,2)
a
```

```python

# finding numbers greater than 50

a>50
```

```python
# now putting the boolean indexing to array to get the resultant output its called boolean masking 
a[a>50]
```

```python



# find the even numbers 
a[a%2==0]
```

```python


# find all numbers greater than 50 and even

a[(a>50) & (a%2==0)]   # we are doing with boolean value so we use bitwise operator
```

```python

# find all numbers which are not divisible by 7
a[a%7!=0]


```

```python




# broadcasting 

# most of the expalnation and broadcasting rule is explained in note

a=np.arange(12).reshape(4,3)

b=np.arange(3).reshape(1,3)


print(a+b)
```

```python




# another example of broadcasting
c=np.arange(12).reshape(4,3)
d=np.arange(3)


print(c+d)
```

```python

 # working with mathematical function 

# sigmoid


# sigmoid value range from 0 to 1

def sigmoid(array):
    return 1/(1+np.exp(-(array)))
a=np.arange(50)

sigmoid(a)
            


```

```python




# mean squared error 


original=np.random.randint(1,100,25)
predicted=np.random.randint(1,100,25)


original
```

```python
predicted
```

```python


def mse(a,p):
    return np.mean((a-p)**2)

mse(original,predicted)
```
