```python
# find top 3 genres by total earnings
movie.groupby('Genre').sum(numeric_only=True).sort_values(by='Gross',ascending=False).head(3)
```

```python


# find the genre with highest avg imdb rating
movie.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python


# director with most popularity
movie.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python



# highest rated movie of each genres
movie.groupby('Genre')['IMDB_Rating'].max()
```

```python
# complex 



# applyong mutiple aggerate functions in different columns using    dict 

genres.agg(
    {
        'Gross':'max',
        'No_of_Votes':'mean'
    }
)
```

```python


# merging both concepts
genres.agg(
    {
        'Gross':['max','min'],
        'No_of_Votes':['mean','max']
    }
)
```

```python
#qns

# top 10 batsman in terms of runs
npl=pd.read_csv(r"C:\Users\acer\Python-Sql-1\numpy_pandas_matplotlib_seaborn_15\dataset_dtaframes\deliveries.csv")
```

```python
npl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# create a function that can return  the highest score of any batsman on his career ball
# 

def high(run):
    return run['batsman_runs'].max()




npl.groupby('batsman').apply(high)
```
