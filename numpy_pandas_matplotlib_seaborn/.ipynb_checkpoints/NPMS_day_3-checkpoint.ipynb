```python

#np.unique
a=np.array([1,1,2,2,3,3,4,5,6,6,7,8,89,0])
np.unique(a)
```

```python


#np.expand_dims()
# change the shape only not data 
# here axis work as row and column change as 0 and 1 but opposite in concatenate 

np.expand_dims(a,axis=1)
```

```python

np.expand_dims(a,axis=0)
```

```python


#np.where()
# find the indices position greater than 50
a=np.array([1,1,2,2,3,3,4,5,6,6,7,8,89,0])
np.where(a>50)
```

```python
# replace the value with 0 if it is grerater than 50

# np.where(condition,true,false)

np.where(a>50,0,a)


```

```python
np.where(a%2==0,0,a)
```

```python



# set functions

#union
n=np.array([1,2,3,4,5,6,8])
p=np.array([3,4,5,6,7,8,9])
np.union1d(n,p)
```

```python

# intersection
np.intersect1d(n,p)
```

```python


# set difference
np.setdiff1d(n,p)
```

```python
np.setdiff1d(p,n)
```

```python

#xor
np.setxor1d(n,p)
```

```python

# setin

np.in1d(n,[1,2,3])
```

```python
#np.clip


# most imp
print(n)
np.clip(n,a_min=3,a_max=4)
```
