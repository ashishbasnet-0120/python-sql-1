

### opening a csv file from a URL

```python

# basically its a code snippet 
# just need to paste the url
# no need to remember all the things 

import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```


### sep and names parameters

#####  basically the read_csv() read the csv because the default value of sep parameter is ',' so if you are working with tsv file 
##### provide sep parameter as "\t"  .although it reads the  tsv file it makes the mistake that it makes 1st row of data to columns
##### to remove this default provide names parameter with list of column names 

```python



pd.read_csv('movie_titles_metadata.tsv',sep='\t')
```

```python
pd.read_csv('movie_titles_metadata.tsv',sep='\t',names=['sn','name','year','rating','votes','genre'])
```


### index_col parameter



###### it is used to remove the default index of pandas if you dont want

```python
pd.read_csv('aug_train.csv')
```

```python
pd.read_csv('aug_train.csv',index_col='enrollee_id')
```

## header paraameter



```python
# treating coln name as row in this case we use header parameter
pd.read_csv('test.csv')
```

```python
pd.read_csv('test.csv',header=1)
```



### use_cols parameter
#### loading the colns which you required


```python
pd.read_csv('aug_train.csv',usecols=['enrollee_id','city','gender'])
```


#### squeeze parameter


```python
pd.read_csv('aug_train.csv',usecols=['gender']).squeeze()
```



#### skiprows and nrows 
##### skiprows skip the rows which you dont want and nrows only loads the data you want


```python
pd.read_csv('aug_train.csv',index_col='enrollee_id')
```

```python
pd.read_csv('aug_train.csv',skiprows=[0,1,2,3])
```

```python
pd.read_csv('aug_train.csv',nrows=50)
```
