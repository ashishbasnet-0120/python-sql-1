```python
# find all the final winners 
winner=ipl[ipl['MatchNumber']=='Final']
winner[['Season','WinningTeam']]
```

```python


# how many  super over finishes have occured
ipl[ipl['SuperOver']=='Y'].shape[0]
```

```python
# how many matches the csk won in kolkata
win=ipl[ipl['City']=='Kolkata']['WinningTeam']=='Chennai Super Kings'
win[win].shape
```

```python
# toss winner is match winner in percentage
(ipl[ipl['TossWinner']==ipl['WinningTeam']].shape[0]/ipl.shape[0])*100
```

```python


# movie with rate higher than 8 and votes >100000
movie[(movie['imdb_rating']>8) & (movie['imdb_votes']>10000)]['title_x']
```
