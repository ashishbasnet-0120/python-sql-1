```python

# moving within a file 10 char then 5 char

with open('sample.txt','r') as f:
    print(f.read(10))
    print(f.read(5))
```

```python



# loading a big file in memory 


big_file=['Ashish Basnet' for i in range(200)]   # list comprehension

with open('big.txt','w') as f:
    f.writelines(big_file)

with open('big.txt','r') as file:
    while True:
        chunk = file.read(100)
        if not chunk:
            break
        print(chunk)
```

```python



# seek and tell function

with open('sample.txt','r') as f:
    print(f.read(5))
    print(f.tell())
    f.seek(0)
    print(f.read(5))


# output is Aaman
```

```python


# seek during write 
with open('demo.txt','w') as f:
    f.write("suman")
    f.seek(0)
    f.write("Aa")
```

```python
# serialization using json module 
# list
import json as jsn
fruits=['Apple','Mango','Banana']
with open('fruits.json','w') as f:
    jsn.dump(fruits,f)


# same can be done with dictionary
```

```python
import json as jsn
fruits=['Apple','Mango','Banana']
with open('fruit.json','w') as f:
    jsn.dump(fruits,f,indent=4)
```

```python

# deserialization

import json

with open('fruits.json','r') as f:
    
    d=json.load(f)
    print(d)
    print(type(d))
```
