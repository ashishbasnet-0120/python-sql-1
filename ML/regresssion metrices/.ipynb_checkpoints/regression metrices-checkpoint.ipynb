```python


import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python

df=pd.read_csv('placement.csv')

df.head()
```

### plotting the scatter plot

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('cgpa')
plt.ylabel('Placement(LPA)')
```

## steps involved in machine learning algorithms 
## splitting x and y 


```python
# x=df[all rows, 1 column]

####      ml model always expect input 2d that is row nad column    so we have to give dataframe not series 
x=df.iloc[:,0].to_frame()
x
```

```python
y=df.iloc[:,-1].to_frame()
y
```

```python




from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python


from sklearn.linear_model import LinearRegression
```

```python


# craeting object of linearregression class


lr=LinearRegression()
```

```python

# testing the data


lr.fit(x_train,y_train)
```

```python
x_train
```

```python
y_train
```

```python


#  sk learn excepts 2d that is rows and column
lr.predict(x_test.iloc[[0]])
```

```python


plt.scatter(df['cgpa'],df['package'])
plt.plot(df['cgpa'],lr.predict(df[['cgpa']]),color='red')
plt.xlabel('cgpa')
plt.ylabel('Placement(LPA)')
```





###  regression metrices 




```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python


y_pred=lr.predict(x_test)
y_pred
```

```python
y_test
```

```python
print('MAE',mean_absolute_error(y_test,y_pred))
```

```python


print('MSE',mean_squared_error(y_pred,y_test))
```

```python


print('RMSE',np.sqrt(mean_squared_error(y_pred,y_test)))
```

```python


print('r2 score', r2_score(y_test,y_pred))
```

```python


# adjusted r2 score


r2=r2_score(y_test,y_pred)
r2
```

```python

x_test.shape
```

```python


adjusted_r2=1-(((1-r2)*(40-1))/(40-1-1))

adjusted_r2
```
